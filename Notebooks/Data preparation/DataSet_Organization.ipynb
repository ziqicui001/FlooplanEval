```python
import os
import re
import pandas as pd

folder_path = '/content/drive/MyDrive/AI_graph/Organized_Dataset02/Graph'

# Get a list of CSV file names sorted numerically
file_names = sorted([file for file in os.listdir(folder_path) if file.endswith('_graphlist.csv')], key=lambda x: int(re.search(r'(\d+)_', x).group(1)))

dataframes = []
for file_name in file_names:
    file_path = os.path.join(folder_path, file_name)
    df = pd.read_csv(file_path)

    # Check if the dataframe is empty and fill it with a row of zeros if necessary
    if df.empty:
        df = pd.DataFrame(columns=df.columns)
        df.loc[0] = [0] * len(df.columns)

    dataframes.append(df)

merged_df = pd.concat(dataframes, ignore_index=True)

merged_df.to_csv('/content/drive/MyDrive/AI_graph/Organized_Dataset02/MergedGraphData/merged05.csv', index=False)
```
